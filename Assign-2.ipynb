```python
import pandas as pd
import numpy as np
import folium

dataframe1=pd.read_excel("coordinates-pk.xlsx")
dataframe2=pd.read_excel("cities-covid-19.xlsx")

#now we clean the data in it
dataframe1.fillna(0)
#we shall remove the unnamed columns from the dataset
missing_data = pd.isna(dataframe1.columns).sum()
print("Missing data of 1st file:"+str(missing_data))

dataframe2.fillna(0)
#we shall remove the unnamed columns from the dataset
missing_data2 = pd.isna(dataframe2.columns).sum()
print("Missing data of 2nd file:"+str(missing_data2))

dataframe=pd.merge(dataframe1,dataframe2)


#loading the map of pakistan
Pakistan=folium.Map(location=[30.375321,69.345116],zoom_start=5.2)


#now we shall add all the data on map of Pakistan
for cities,latitude,longitude,cases,Population in zip(list(dataframe['Cities']),list(dataframe
['Latitude']),list(dataframe['Longitude']),list(dataframe['Number-of-confirmed-cases']),list(dataframe['Population'])):
    folium.CircleMarker(location=[latitude,longitude],
                       radius=6,
                       color='green',
                       fill= True,
                       fill_color='green').add_to(Pakistan)
    
    #now for displaying
    folium.Marker(location=[latitude,longitude],
                  popup=folium.Popup(('<strong><b>Cities : '+cities+'</strong> <br>' +
                  '<strong><b>Latitude : '+str(latitude)+'</strong> <br>' +
                  '<strong><b>Longitude: '+str(longitude)+'</strong> <br>' +
                  '<strong><b>Number-of-confirmed-cases : '+str(cases)+'</strong> <br>' +
                  '<strong><b>Population : '+str(Population)+'</strong>'),max_width=200)).add_to(Pakistan)



```
